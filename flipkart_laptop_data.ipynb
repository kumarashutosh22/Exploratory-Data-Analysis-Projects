```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&marketplace=FLIPKART&sort=price_desc&page=1'

req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content,'lxml')

print(soup.prettify())
```
